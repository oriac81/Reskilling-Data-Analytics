#### Exercici 1: Calculadora de l'índex de massa corporal
Escriu una funció que calculi l'IMC ingressat per l'usuari/ària, és a dir, qui ho executi haurà d'ingressar aquestes dades. Pots obtenir més informació del seu càlcul en:

Índice de masa corporal IMC que es y como se calcula.
La funció ha de classificar el resultat en les seves respectives categories

##### Solució. Calculadora de IMC

```python
""" Calculadora de IMC """

# Llista de categories (llindar superior exclusiu, missatge)
CATEGORIES = [
    (18.5, 'un pes baix.'),
    (25, 'un pes normal.'),
    (30, 'sobrepes.'),
    (35, 'obesitat grau I.'),
    (float('inf'), 'obesitat grau II.')
]

def obtenir_dada(missatge, min_val, max_val):
    """Demana i valida una dada dins dels límits especificats."""
    while True:
        try:
            valor = float(input(missatge))
            if min_val < valor <= max_val:
                return valor
            print(f"Error: Valor entre {min_val} i {max_val}")
        except ValueError:
            print("Error: Introdueix un número vàlid")

def demanar_dades():
    """Obté alçada i pes vàlids."""
    alçada = obtenir_dada("Alçada en m (Ex: 1.75): ", 0, 2.5)
    pes = obtenir_dada("Pes en kg (Ex: 68.5): ", 0, 400)
    return alçada, pes

def calculadora_imc(alçada, pes):
    """Calcula l'Índex de Massa Corporal."""
    return round(pes / (alçada ** 2), 1)

def classificar_imc(imc):
    """Determina la categoria utilitzant cerca eficient."""
    for llindar, missatge in CATEGORIES:
        if imc < llindar:
            return missatge

def mostrar_resultat(alçada, pes, imc, categoria):
    """Mostra resultat detallat amb f-string."""
    print(
        f"Amb una alçada de {alçada}m i un pes de {pes}kg, el teu IMC és {imc} i correspon a {categoria}"
    )

def main():
    """Flux principal del programa."""
    alçada, pes = demanar_dades()
    imc = calculadora_imc(alçada, pes)
    categoria = classificar_imc(imc)
    mostrar_resultat(alçada, pes, imc, categoria)

#Executar el programa
main()
```

#### Exercici 2: Convertidor de temperatures
Existeixen diverses unitats de temperatura utilitzades en diferents contextos i regions. Les més comunes són Celsius (°C), Fahrenheit (°F) i Kelvin (K). També existeixen altres unitats com Rankine (°Ra) i Réaumur (°Re). Selecciona almenys 2 conversors, de tal manera que en introduir una temperatura retorni, com a mínim, dues conversions.

##### Solució

```python
"""
Conversor de temperatures amb suport per a Celsius, Fahrenheit, Kelvin, Rankine i Réaumur
"""

# Constants globals
MINIMS = {
    'C': -273.15,
    'F': -459.67,
    'K': 0.0,
    'RA': 0.0,
    'RE': -218.52
}

UNITATS = ['C', 'F', 'K', 'RA', 'RE']

UNIT_INFO = {
    'C': {'symbol': '°C', 'name': 'Celsius'},
    'F': {'symbol': '°F', 'name': 'Fahrenheit'},
    'K': {'symbol': 'K', 'name': 'Kelvin'},
    'RA': {'symbol': '°Ra', 'name': 'Rankine'},
    'RE': {'symbol': '°Re', 'name': 'Réaumur'}
}

def obtenir_unitat() -> str:
    """Demana i valida la unitat de temperatura.
    
    Returns:
        str: Unitat vàlida introduïda (C, F, K, RA, RE)
    """
    while True:
        prompts = []
        for unit in UNITATS:
            prompts.append(f"{unit} per a {UNIT_INFO[unit]['name']}")
        prompt = "Introdueix la unitat de temperatura (" + ", ".join(prompts) + "): "
        unitat = input(prompt).upper()
        if unitat in UNITATS:
            return unitat
        print("Error: Unitat no vàlida. Tria:", ", ".join(UNITATS))

def obtenir_temperatura(unitat: str) -> float:
    """Demana i valida la temperatura segons la unitat.
    
    Args:
        unitat (str): Unitat de temperatura vàlida
    
    Returns:
        float: Temperatura vàlida introduïda
    """
    while True:
        try:
            temp = float(input('Introdueix la temperatura: '))
            if temp < MINIMS[unitat]:
                print(f"Error: Temperatura inferior al zero absolut ({MINIMS[unitat]} {UNIT_INFO[unitat]['symbol']})")
                continue
            return temp
        except ValueError:
            print("Error: Introdueix un valor numèric vàlid (ex: 25.3)")

def convertir_a_celsius(unitat: str, temp: float) -> float:
    """Converteix qualsevol unitat a Celsius.
    
    Args:
        unitat (str): Unitat original
        temp (float): Temperatura original
    
    Returns:
        float: Temperatura convertida a Celsius
    """
    if unitat == "C":
        return temp
    if unitat == "F":
        return (temp - 32) * 5/9
    if unitat == "K":
        return temp - 273.15
    if unitat == "RA":
        return (temp - 491.67) * 5/9
    if unitat == "RE":
        return temp * 5/4
    raise ValueError("Unitat no reconeguda")

def calcular_conversions(celsius: float) -> dict:
    """Calcula totes les conversions a partir de Celsius.
    
    Args:
        celsius (float): Temperatura en Celsius
    
    Returns:
        dict: Diccionari amb totes les conversions
    """
    return {
        'C': celsius,
        'F': celsius * 9/5 + 32,
        'K': celsius + 273.15,
        'RA': (celsius + 273.15) * 9/5,
        'RE': celsius * 4/5
    }

def mostrar_resultats(unitat_original: str, temp_original: float, resultats: dict) -> None:
    """Mostra els resultats formatats.
    
    Args:
        unitat_original (str): Unitat d'entrada
        temp_original (float): Temperatura original
        resultats (dict): Diccionari amb totes les conversions
    """
    original_symbol = UNIT_INFO[unitat_original]['symbol']
    original_name = UNIT_INFO[unitat_original]['name']
    print(f"\n{temp_original}{original_symbol} ({original_name}) equivalen a:")
    for unit in UNITATS:
        if unit == unitat_original:
            continue
        valor = resultats[unit]
        unit_name = UNIT_INFO[unit]['name']
        print(f"- {valor:.2f} {unit_name}")

def main() -> None:
    """Funció principal que gestiona el flux del programa."""
    unitat = obtenir_unitat()
    temperatura = obtenir_temperatura(unitat)
    celsius = convertir_a_celsius(unitat, temperatura)
    conversions = calcular_conversions(celsius)
    mostrar_resultats(unitat, temperatura, conversions)

# Executar el programa
main()
```

#### Exercici 3: Comptador de paraules d'un text

Escriu una funció que donat un text, mostri les vegades que apareix cada paraula.

##### Solució 1. Amb COUNTER exclusivament

```python
# Importa la classe Counter per a recomptes eficients
from collections import Counter

# Constants globals (MAJÚSCULES)
CARACTERS_NETEJA = ',.!¿?¡()[]{}:;"\'’“”«»—'  # Tots els caràcters no alfabètics a eliminar

# Funció principal que coordina el flux
def main():
    # 1. Obtenir entrada d'usuari
    text = input("Introdueix text: ")
    
    # 2. Netejar text: minúscules i eliminar símbols
    text_net = text.lower().translate(str.maketrans('', '', CARACTERS_NETEJA))
    
    # 3. Tokenitzar en paraules individuals
    paraules = text_net.split()
    
    # 4. Recomptar freqüències amb Counter
    comptador = Counter(paraules)
    
    # 5. Ordenar alfabèticament els resultats
    resultat_ordenat = sorted(comptador.items())
    
    # 6. Mostrar resultats formatats
    print("\nRecompte alfabètic de paraules:")
    for paraula, quantitat in resultat_ordenat:
        print(f"- {paraula}: {quantitat} cop{'s' if quantitat > 1 else ''}")

# Executar el programa

main()
```

##### Solució 2. Implementació manual d'un sistema Map-Reduce per a recompte de paraules


```python
"""
Implementació MapReduce per a recompte de paraules amb proposit educatiu.
Aquest codi il·lustra el procés de MapReduce en Python, incloent les fases de mapeig, agrupació i reducció.     
"""

# Constants globals (MAJÚSCULES)
CARACTERS_NETEJA = ',.!¿?¡()[]{}:;"\'’“”«»—'

# Fase 1: Mapa (Transformació)
def map_paraula(text: str) -> list[tuple[str, int]]:
    """Genera tuples (paraula, 1) per a cada ocorrència.
    
    Args:
        text (str): Text brut d'entrada
        
    Returns:
        Llista de tuples (paraula_netejada, 1)
    """
    text_net = text.lower().translate(str.maketrans('', '', CARACTERS_NETEJA))
    return [(paraula, 1) for paraula in text_net.split()]

# Fase 2: Shuffle (Agrupació)
def shuffle_paraules(maped: list[tuple[str, int]]) -> dict[str, list[int]]:
    """Agrupa els valors per clau (paraula).
    
    Args:
        maped: Llista de tuples (paraula, 1) del mapper
        
    Returns:
        Diccionari {paraula: [1, 1, ...]}
    """
    shuffled = {}
    for paraula, compte in maped:
        shuffled.setdefault(paraula, []).append(compte)
    return shuffled

# Fase 3: Redueix (Agregació)
def reduce_paraula(shuffled: dict[str, list[int]]) -> dict[str, int]:
    """Suma els valors agrupats per cada paraula.
    
    Args:
        shuffled: Diccionari amb llistes de comptes
        
    Returns:
        Diccionari final {paraula: total}
    """
    return {paraula: sum(comptes) for paraula, comptes in shuffled.items()}

# Funció principal
def main():
    """Coordina tot el flux MapReduce."""
    # 1. Entrada
    text = input("Introdueixi el seu text: ")
    
    # 2. Execució fase Map
    mapped = map_paraula(text)
    
    # 3. Execució fase Shuffle
    shuffled = shuffle_paraules(mapped)
    
    # 4. Execució fase Reduce
    resultat = reduce_paraula(shuffled)
    
    # 5. Ordenar i mostrar resultats
    print("\nRecompte alfabètic de paraules:")
    for paraula, quantitat in sorted(resultat.items()):
        print(f"- {paraula}: {quantitat} cop{'s' if quantitat > 1 else ''}")

# Executar el programa
main()
```

Els valors i claus en el diccionari original són únics; si aquest no és el cas, la funció hauria d'imprimir un missatge d'advertiment.



reverse_dictionary ({'a': 1, 'b': 2, 'c': 3})

##### Solució 1

```python
def reverse_dictionary(diccionari):
    """Inverteix un diccionari verificant valors duplicats."""
    valors_vistos = set()  # Set per a millor rendiment
    for valor in diccionari.values():
        if valor in valors_vistos:
            return "Error: multiple keys for one value"
        valors_vistos.add(valor)
    
    return {valor: clau for clau, valor in diccionari.items()}

def mostrar_resultat(num_test, resultat):
    """Mostra els resultats formatejats."""
    print(f"Resultat {num_test}: {resultat}")

def main():
    """Funció principal amb casos de prova."""
    # Llista de diccionaris per testejar
    tests = [
        {'a': 1, 'b': 2, 'c': 3},
        {'x': 'apple', 'y': 'banana', 'z': 'banana'}
    ]
    
    # Executar tests i mostrar resultats
    for i, test in enumerate(tests, 1):
        resultat = reverse_dictionary(test)
        mostrar_resultat(i, resultat)

# Executar el programa
main()
```

##### Solució 2

```python
"""
Inversor de diccionaris amb control de valors duplicats
"""

def reverse_dictionary(diccionari: dict) -> dict | str:
    """Inverteix un diccionari verificant valors únics.
    
    Args:
        diccionari (dict): Diccionari original a invertir
        
    Returns:
        dict: Diccionari invertit si és possible
        str: Missatge d'error si hi ha valors duplicats
    
    Raises:
        ValueError: Si el paràmetre no és un diccionari
    """
    if not isinstance(diccionari, dict):
        raise ValueError("L'entrada ha de ser un diccionari")
    
    valors_vistos = set()
    for valor in diccionari.values():
        if valor in valors_vistos:
            return "Error: multiple keys for one value"
        valors_vistos.add(valor)
    
    return {valor: clau for clau, valor in diccionari.items()}

def main() -> None:
    """Funció principal amb casos de prova."""
    # Casos de prova
    tests = [
        {'a': 1, 'b': 2, 'c': 3},
        {'x': 'poma', 'y': 'banana', 'z': 'banana'},
        "això no és un diccionari"  # Prova d'error
    ]
    
    # Executar i mostrar resultats
    for i, test in enumerate(tests, 1):
        try:
            resultat = reverse_dictionary(test)
            print(f"\nTest {i} ({test}):")
            print(f"Resultat: {resultat}")
        except ValueError as e:
            print(f"\nTest {i} ERROR: {e}")

# Executar el programa
main()
```

#### Exercici 1: Diccionari invers amb duplicats

Continuant amb l'exercici 4 del nivell 1: al client es va oblidar de comentar un detall i resulta que els valors en el diccionari original poden duplicar-se i més, per la qual cosa les claus intercanviades poden tenir duplicats. En aquest cas, en l'exercici anterior imprimies un missatge d'advertiment, ara, els valors del diccionari resultant hauran d'emmagatzemar-se com una llista. Tingues en compte que si és un valor únic no ha de ser una llista.


reverse_dictionary_plus({'x': 'apple', 'y': 'banana', 'z': 'banana'})

{'apple': 'x', 'banana': ['y', 'z']}

##### Solució

```python
def reverse_dictionary_plus(diccionari):
    """
    Inverteix un diccionari gestionant valors duplicats
    
    Args:
        diccionari (dict): Diccionari original amb claus:valors
    
    Returns:
        dict: Diccionari invertit amb valors:claus o llistes de claus
    """
    # Fase 1: Agrupació de claus amb el mateix valor
    agrupacio = {}
    for clau, valor in diccionari.items():
        if valor not in agrupacio:
            agrupacio[valor] = []
        agrupacio[valor].append(clau)
    
    # Fase 2: Simplificació de llistes unitàries
    resultat = {}
    for valor, claus in agrupacio.items():
        resultat[valor] = claus if len(claus) > 1 else claus[0]
    
    return resultat

def main():
    """Executa un conjunt de proves i mostra els resultats"""
    # Definim les proves com a llista de tuples (entrada, sortida_esperada)
    casos_prova = [
        ({'a': 1, 'b': 2, 'c': 3}, {1: 'a', 2: 'b', 3: 'c'}),
        ({'x': 'apple', 'y': 'banana', 'z': 'banana'}, {'apple': 'x', 'banana': ['y', 'z']}),
        ({'a': 'fruit', 'b': 'fruit', 'c': 'fruit'}, {'fruit': ['a', 'b', 'c']})
    ]
    
    for i, (entrada, esperat) in enumerate(casos_prova, 1):
        obtingut = reverse_dictionary_plus(entrada)
        estat = "✓" if obtingut == esperat else "✗"
        
        print(f"Prova {i}:")
        print(f"Entrada:   {entrada}")
        print(f"Esperat:   {esperat}")
        print(f"Obtingut:  {obtingut}")
        print(f"Resultat:  {estat}\n")

# Executar el programa
main()
```

El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir en flotants i l'altra on estan els elements que no es van poder convertir. Exemple de la llista que rep el client: [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]


conversion([ '1.3', 'one', '1e10', 'seven', '3-1/2', ('2',1,1.4, 'not-a-number'), [1,2,'3','3.4']])

([1.3, 10000000000.0, 2.0, 1.0, 1.4, 1.0, 2.0, 3.0, 3.4],
['one', 'seven', '3-1/2', 'not-a-number'])

##### Solució 1

```python
"""
Processament de dades niuades amb conversió a float
"""

from typing import Union, List, Tuple

def processar_element(element: Union[list, tuple, object], convertits: List[float], no_convertits: List[object]) -> None:
    """Processa recursivament elements niuats intentant convertir a float."""
    if isinstance(element, (list, tuple)):
        for subelement in element:
            processar_element(subelement, convertits, no_convertits)
    else:
        try:
            convertits.append(float(element))
        except (ValueError, TypeError):
            no_convertits.append(element)

def processar_dades(dades: Union[list, tuple]) -> Tuple[List[float], List[object]]:
    """Orquestra el processament de dades niuades."""
    convertits = []
    no_convertits = []
    for element in dades:
        processar_element(element, convertits, no_convertits)
    return convertits, no_convertits

def mostrar_resultats(convertits: List[float], no_convertits: List[object]) -> None:
    """Mostra els resultats del processament."""
    print(f"Convertits ({len(convertits)}): {convertits}")
    print(f"No convertits ({len(no_convertits)}): {no_convertits}")

def main() -> None:
    """Executa el flux principal amb dades de prova."""
    entrada = [
        '1.3', 'one', '1e10', 'seven', '3-1/2',
        ('2', 1, 1.4, 'not-a-number'),
        [1, 2, '3', '3.4']
    ]
    convertits, no_convertits = processar_dades(entrada)
    mostrar_resultats(convertits, no_convertits)

# Executar el programa
main()
```

##### Solució 2

```python
"""
Processador de dades niuades amb conversió optimitzada a float
"""

from typing import Union, List, Tuple

def processar_dades(dades: Union[List, Tuple]) -> Tuple[List[float], List[object]]:
    """
    Processa estructures de dades niuades i classifica elements convertibles
    
    Args:
        dades: Estructura niuada (llistes/tuples) amb elements diversos
    
    Returns:
        (llista_valors_convertits, llista_valors_no_convertits)
    """
    convertits = []
    no_convertits = []

    def _processar_element(element: Union[list, tuple, object]) -> None:
        """Funció recursiva interna per processar elements."""
        if isinstance(element, (list, tuple)):
            for item in element:
                _processar_element(item)
        else:
            try:
                convertits.append(float(element))
            except (TypeError, ValueError):
                no_convertits.append(element)

    for element in dades:
        _processar_element(element)

    return convertits, no_convertits

def mostrar_resultats(convertits: List[float], no_convertits: List[object]) -> None:
    """Mostra resultats amb el format específic sol·licitat."""
    print(f"✅ CONVERTITS ({len(convertits)}): {convertits}")
    print(f"❌ NO CONVERTITS ({len(no_convertits)}): {no_convertits}")

def main() -> None:
    """Flux principal amb casos de prova."""
    entrada = [
        '1.3', 'one', '1e10', 'seven', '3-1/2',
        ('2', 1, 1.4, 'not-a-number'),
        [1, 2, '3', '3.4']
    ]
    
    c, nc = processar_dades(entrada)
    mostrar_resultats(c, nc)

# Executar el programa
main()
```

##### Solució 1. Amb COUNTER exclusivament

```python
"""
Analitzador de textos amb recompte de paraules per lletra inicial
"""

from collections import Counter
import re
from typing import Dict, List, Union

# Constants globals
RUTA_FITXER = r"G:\Mi unidad\IT ACADEMY\Reskilling Data Analytics\SPRINT 7\tu_me_quieres_blanca.txt"
CARACTERS_NETEJA = r"[.,!?;:¿¡\"'()–—«»\n\r\t]"

def llegir_fitxer(nom_fitxer: str) -> str:
    """Llegeix un fitxer de text amb gestió d'errors.
    
    Args:
        nom_fitxer (str): Ruta completa del fitxer
        
    Returns:
        str: Contingut del fitxer o cadena buida en cas d'error
    """
    try:
        with open(nom_fitxer, 'r', encoding='utf-8') as f:
            return f.read()
    except FileNotFoundError:
        print(f"Error: El fitxer '{nom_fitxer}' no existeix")
    except Exception as e:
        print(f"Error inesperat en llegir el fitxer: {str(e)}")
    return ""

def processar_text(text: str) -> List[str]:
    """Neteja i tokenitza el text en paraules.
    
    Args:
        text (str): Text brut a processar
        
    Returns:
        List[str]: Llista de paraules netes
    """
    text_net = re.sub(CARACTERS_NETEJA, ' ', text.lower())
    return [paraula for paraula in text_net.split() if paraula]

def comptar_paraules(paraules: List[str]) -> Counter:
    """Recompta freqüència de paraules.
    
    Args:
        paraules (List[str]): Llista de paraules
        
    Returns:
        Counter: Objecte amb recomptes
    """
    return Counter(paraules)

def agrupar_per_inicial(recompte: Counter) -> Dict[str, Dict[str, int]]:
    """Agrupa paraules per lletra inicial amb recompte.
    
    Args:
        recompte (Counter): Recompte de paraules
        
    Returns:
        Dict[str, Dict[str, int]]: Diccionari agrupat per inicial
    """
    grups = {}
    for paraula, freq in recompte.items():
        if not paraula:
            continue
        
        lletra = paraula[0].lower()
        if lletra.isalpha():
            grups.setdefault(lletra, {})[paraula] = freq
    return grups

def ordenar_diccionari(grups: Dict[str, Dict[str, int]]) -> Dict[str, Dict[str, int]]:
    """Ordena diccionari alfabèticament.
    
    Args:
        grups (Dict): Diccionari per ordenar
        
    Returns:
        Dict: Diccionari ordenat
    """
    return {
        lletra: dict(sorted(paraules.items()))
        for lletra, paraules in sorted(grups.items())
    }

def generar_estructura(nom_fitxer: str) -> Dict[str, Dict[str, int]]:
    """Orquestra tot el flux de processament.
    
    Args:
        nom_fitxer (str): Ruta del fitxer
        
    Returns:
        Dict: Estructura final ordenada
    """
    contingut = llegir_fitxer(nom_fitxer)
    if not contingut:
        return {}
    
    paraules = processar_text(contingut)
    recompte = comptar_paraules(paraules)
    return ordenar_diccionari(agrupar_per_inicial(recompte))

def imprimir_resultats(estructura: Dict[str, Dict[str, int]]) -> None:
    """Mostra resultats amb format estructurat.
    
    Args:
        estructura (Dict): Diccionari amb dades
    """
    print("{\n" + ",\n".join(
        f"    '{lletra}': {{\n" + 
        ",\n".join(f"        '{paraula}': {freq}" for paraula, freq in paraules.items()) + 
        "\n    }" 
        for lletra, paraules in estructura.items()
    ) + "\n}")

def main() -> None:
    """Flux principal d'execució."""
    estructura = generar_estructura(RUTA_FITXER)
    imprimir_resultats(estructura)

#Executar el programa
main()
```

##### Solució 2. Amb MapReduce

```python
"""
Implementació Map-Reduce per a anàlisi de textos amb agrupació per lletra inicial. Complementem amb COUNTER 
per a gestionar el recompte de paraules de manera eficient.
El codi llegeix un fitxer de text, neteja el contingut, compta les paraules i les agrupa per lletra inicial.    
"""

from collections import Counter
import re
from typing import Dict, List

# Constants globals
RUTA_FITXER = r"G:\Mi unidad\IT ACADEMY\Reskilling Data Analytics\SPRINT 7\tu_me_quieres_blanca.txt"
CARACTERS_NETEJA = r'[.,!?;:¿¡"\'()–—«»\n\r\t]'

def llegir_fitxer(nom_fitxer: str) -> str:
    """Llegeix contingut del fitxer amb gestió d'errors."""
    try:
        with open(nom_fitxer, 'r', encoding='utf-8') as f:
            return f.read()
    except FileNotFoundError:
        print(f"Error: Fitxer '{nom_fitxer}' no trobat")
        return ""
    except Exception as e:
        print(f"Error llegint fitxer: {str(e)}")
        return ""

def map_paraula(text: str) -> List[str]:
    """Fase Map: Neteja i tokenitza el text."""
    text_net = re.sub(CARACTERS_NETEJA, ' ', text.lower())
    return text_net.split()

def reduce_paraula(paraules: List[str]) -> Counter:
    """Fase Reduce: Compta freqüències de paraules."""
    return Counter(paraules)

def agrupa_i_ordena(recompte: Counter) -> Dict[str, Dict[str, int]]:
    """Agrupa per lletra inicial i ordena resultats."""
    grups = {}
    
    # Fase de Shuffle i Ordenació
    for paraula, freq in recompte.items():
        if not paraula:
            continue
        
        lletra = paraula[0].lower()
        if lletra.isalpha():
            grups.setdefault(lletra, {})[paraula] = freq
    
    # Ordenació final
    return {
        lletra: dict(sorted(paraules.items()))
        for lletra, paraules in sorted(grups.items())
    }

def imprimir_resultats(estructura: Dict[str, Dict[str, int]]) -> None:
    """Mostra resultats en format estructurat."""
    print("{\n" + "\n".join(
        f"    '{lletra}': {{\n" + 
        ",\n".join(f"        '{k}': {v}" for k, v in paraules.items()) +
        "\n    }" 
        for lletra, paraules in estructura.items()
    ) + "\n}")

def main():
    """Flux principal Map-Reduce."""
    # Fase d'execució
    text = llegir_fitxer(RUTA_FITXER)
    if not text:
        return
    
    # Pipeline de processament
    paraules = map_paraula(text)
    recompte = reduce_paraula(paraules)
    resultat = agrupa_i_ordena(recompte)
    
    # Sortida formatada
    imprimir_resultats(resultat)

# Executar el programa
main()
```
